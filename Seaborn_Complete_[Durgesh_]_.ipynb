# Visualize Distributions With Seaborn

Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

```python
import seaborn as sns #Seaborn is the only library we need to import for this simple example
import matplotlib.pyplot as plt
```

```python
sns.set_theme() # Apply the default theme
```

```python
tips=sns.load_dataset('tips') # Load an example dataset
```

```python
tips
```

### Line chart

```python
sns.lineplot(data=tips,x='total_bill',y='tip',color='r')
```

### Count chart

```python
sns.countplot(data=tips,x='time')
#it show the the count of each categorical data
```

### BAr plot

```python
ir=sns.load_dataset('iris')
ir.head()
```

```python
sns.barplot(data=ir,x='species',y='sepal_length')
```

```python
sns.pairplot(ir,hue='species')
```

```python
import numpy as np
x=np.linspace(0,10,30)
y=np.sin(x)
sns.scatterplot(x,y,marker='*',color='blue')
```

```python
x=np.array([27,12,4,15,35,62,14,5,9,45,85,8,9,69,45,55])
y=np.array([10,20,30,40,50,60,70,80,90])
sns.distplot(x,y)
```

#### visualization

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
dots=sns.load_dataset('dots')
dots.head()
```

```python
sns.relplot(
    data=dots,kind='line',
    x='time',y='firing_rate'
    ,col='align',hue='choice',
    size='coherence',style='choice',
    facet_kws=dict(sharex=False),
)
```

## Similar functions for similar tasks


```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.kdeplot(data=penguins,x='flipper_length_mm',hue="species", multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

### Figure-level vs. axes-level functions

```python
sns.displot(data=penguins,x="flipper_length_mm", hue="species", multiple='stack')
```

```python
sns.displot(data=penguins,x="flipper_length_mm", hue="species",multiple='stack',kind='kde')
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

### Axes-level functions make self-contained plots

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```
